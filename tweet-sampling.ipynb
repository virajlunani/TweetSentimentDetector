```python
import pandas as pd
import os
```

```python
import zipfile
```

```python
directory = "/content/drive/MyDrive/498/498-project/covid-tweets/"
```

```python
for c_dir in [x for x in os.listdir(directory) if 'sampled' not in x]:
    if c_dir+'_sampled' not in os.listdir(directory):
      os.mkdir(directory+c_dir+'_sampled/')
    for file in os.listdir(directory+c_dir):
      if file[:-4]+'_sampled.csv' not in os.listdir(directory+c_dir+'_sampled/'):
        if 'zip' in file:
          with zipfile.ZipFile(directory+c_dir+'/'+file,"r") as zip_ref:
            zip_ref.extractall(directory+c_dir)
          df = pd.read_csv(directory+c_dir+'/'+file[:-4]+'.csv',header=None)
        elif 'csv' in file:
          df = pd.read_csv(directory+c_dir+'/'+file,header=None)
        else:
          continue
        sample = df.sample(100)
        sample.to_csv(directory+c_dir+'_sampled/'+file[:-4]+'_sampled.csv',
                        header=None,index=False)
```
